# Data_Quest_Analyze_The_Hour_Of_The_Most_Comments
Goal: Indetify the hour that is highly likely to get comments

```python
from csv import reader
opened_file = open('hacker_news.csv')
read_file = reader(opened_file)
hn = list(read_file)
header = hn[0]
hn = hn[1:]
print(header)
print(hn[0])
```

Classify title: Starts with "Ask hn", "Show hn", or other

```python
ask_posts = []
show_posts = []
other_posts = []

for row in hn:
    title = row[1]
    title = title.lower()
    if title.startswith("ask hn"):
        ask_posts.append(row)
    elif title.startswith("show hn"):
        show_posts.append(row)
    else:
        other_posts.append(row)

print(len(ask_posts))
print(len(show_posts))
print(len(other_posts))
```

Category Comment number calculation

```python
total_ask_comments = 0
row_num = 0
for row in ask_posts:
    total_ask_comments += int(row[4])
    row_num += 1
avg_ask_comments = total_ask_comments / row_num
print(avg_ask_comments)

total_show_comments = 0
row_num = 0
for row in show_posts:
    total_show_comments += int(row[4])
    row_num += 1
avg_show_comments = total_show_comments / row_num
print(avg_show_comments)
```

"Ask hm" posts received more comments

Time Comment number calculation

```python
import datetime as dt

result_list = []
for row in ask_posts:
    result_list.append((row[6],int(row[4])))
    
counts_by_hour = {}
comments_by_hour = {}
for row in result_list:
    row_date = dt.datetime.strptime(row[0], "%m/%d/%Y %H:%M")
    row_hour = dt.datetime.strftime(row_date, "%H")
    if row_hour not in counts_by_hour:
        counts_by_hour[row_hour] = 1
        comments_by_hour[row_hour] = row[1]
    elif row_hour in counts_by_hour:
        counts_by_hour[row_hour] += 1
        comments_by_hour[row_hour] += row[1]
    
avg_by_hour = []
for key in comments_by_hour:
    avg_by_hour.append([key, comments_by_hour[key]/counts_by_hour[key]])

swap_avg_by_hour = []
for element in avg_by_hour:
    swap_avg_by_hour.append([element[1],element[0]])

sorted_swap = sorted(swap_avg_by_hour,reverse=True)

for element in sorted_swap[0:5]:
    format_time = dt.datetime.strptime(element[1],"%H").strftime("%H:%M")
    format_string = "{}: {:.2f} average comments per post".format(format_time,float(element[0]))
    print(format_string)
```

Result: 15:00 has the highest change of getting comments
